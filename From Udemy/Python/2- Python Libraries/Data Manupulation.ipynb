```python
df_filled = df.fillna(0)
# print(df_filled)
```

```python
df["Sales_fillna"]= df["Sales"].fillna(df["Sales"].mean())
# print(df)
```

```python
df = df.rename(columns={"Date":"Sales Date"})
df.head()
```

```python
df["newValue"]=df["Value"].fillna(df["Value"].mean()).astype(int)
# print(df)
```

```python
df["SqNewValue"]= df["Value"].apply(lambda x:x*2)
# print(df.tail())
# df.head()
```

```python
grouped_mean = df.groupby("Product")["Value"].mean()
# print(grouped_mean)
```

```python
gM = df.groupby("Region")["Value"].mean()
# print(gM)
```

```python
gM = df.groupby("Region")["Value"].sum()
# print(gM)
```

```python
twoGroupedSum = df.groupby(["Product","Region"])["Value"].sum()
# print(twoGroupedSum)
```

```python
gm = df.groupby("Region")["Value"].agg(["mean","sum","count"])
# print(gm)
```
