### 1. Introduction

Air pollution is a major environmental issue in London, affecting health and quality of life.  
This project explores how weather conditions such as **temperature**, **humidity**, and **wind speed** influence **PM2.5** air pollution levels in London (2023‚Äì2024).  
Data used here are sample datasets simulating realistic daily weather and pollution trends.


## 2. Data Loading and Overview

```

import pandas as pd

# Load data
air = pd.read_csv("london_air_quality.csv")
weather = pd.read_csv("london_weather.csv")

# Convert to datetime
air['date'] = pd.to_datetime(air['date'])
weather['date'] = pd.to_datetime(weather['date'])

# Merge
df = pd.merge(air, weather, on='date')
df.head()
```

## 3. Exploratory Data Analysis (EDA)

```

import matplotlib.pyplot as plt
import seaborn as sns

# Correlation heatmap
corr = df[['PM2.5', 'temperature', 'humidity', 'wind_speed']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Between Air Quality and Weather")
plt.show()

# Pairplot visualization
sns.pairplot(df[['PM2.5', 'temperature', 'humidity', 'wind_speed']])
plt.show()
```

## 4. Regression Modelling

```

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error

# Model setup
X = df[['temperature', 'humidity', 'wind_speed']]
y = df['PM2.5']

model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

print("R¬≤:", r2_score(y, y_pred))
print("RMSE:", mean_squared_error(y, y_pred, squared=False))

pd.DataFrame({'Variable': X.columns, 'Coefficient': model.coef_})
```
